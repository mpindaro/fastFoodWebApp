Un ristorante è formato da:
- username
- password
- nome ristorante
- Telefono
- partitaIVA
- indirizzo
- recensioni

```python
import json
```

```python
ristoratori = []
for i in range(0,3):
    recensioni = []
    username = input("username")
    psw = input("psw")
    nomeRistorante = input("nome")
    telefono = input("telefono")
    iva= input("iva")
    indirizzo = input("indirizzo")
    for j in range (0,2):
        stelle=input("stelle")
        testo=input("recensioni")
        recensioni.append({'stelle': stelle,'testo': testo})
    ristoratori.append({'username': username, 'password': psw, 'nome':nomeRistorante, 'telefono':telefono, 'partitaIva': iva, 'recensioni':recensioni})
```

```python
ristoratori
```

```python
with open("menu.json","r") as menufile:
    menu=json.load(menufile)
```

```python
ristoratori[0]['menu'] = [piatto['ID'] for piatto in menu]
ristoratori[1]['menu'] = [piatto['ID'] for piatto in menu \
                          if piatto['categoria'] not in ['Croissants','Frosty','Biscuits']]
ristoratori[2]['menu'] = [piatto['ID'] for piatto in menu]
```

```python
with open("ristoranti.json", "w") as writefile:
    json.dump(ristoratori, writefile, indent=1)
```
